# CitiBike Analysis

### Project Overview

The project is all about the analysis being done on New York Citi Bike data, where we have used the data for the month of August 2019 to analyse the bike sharing business scope and use the same for pitching to the investors. Here we have used data visualization tools to present the results to the investors. 

We have done the following steps to fulfill the aforementioned purpose:
1. Some data formatting and cleaning by using Pandas
2. Importing the prepared data into Tableau
3. Create dashboard in Tableau (https://public.tableau.com/app/profile/gideon.nyamaah7372/viz/NycCitiBikeAnalysis_16768792510170/NYCCitiBikeAnalysis?publish=yes)
4. Present the story coming out of the data

### Summary
The data tells the demand of the nyc citibike in the month of august. The riders are spread across all the age groups. It demand seems quite high during the peak office hours in the morning time 7am to 9am and also during the evening 4pm to 7pm. This may be because of the people going office in the morning and returning in the evening so these bikes are an alternative to public and personal vehicle. The majority of the rides were in the very busy Manhattan Island, taken by male users during morning and evening rush hours.
